<h2>
    Determinação da Temperatura Adiabática
</h2>

O método de determinação da temperatura adiabática de chama se fundamenta na primeira lei da termodinâmica e em sua aplicação a sistemas reativos, principalmente no caso da combustão, de forma tal que os conceitos e relações termodinâmicas precedentes e decorrrentes da aplicação desta lei ao estudo da chama devem ser bem conhecidos.

<h3>
    Primeira Lei da Termodinâmica e Entalpia
</h3>

A primeira lei da termodiâmica equaciona as diferenças formas de energia manifestas em transformações térmicas, sejam de natureza física ou química e pode ser escrita  na forma da Eq. \ref{Eq:primeiralei}, onde $dU$ é a mudança na energia interna, $P dV$ o trabalho devido à mudanças no volume do sistema e $\delta q$ o calor trocado com a vizinhança. A notação diferencial distinta usada para $Q$ se deve ao fato do processo ser dependente das variáveis de estado como $U$, $P$ e $V$. 

\begin{equation}
\label{Eq:calordereacaoeentalpia}
dU + p dV = \delta Q
\end{equation}

Da definição é possível observar que qualquer absorção ou perda energética na forma de calor ($dQ$) pode manifestar-se como energia interna $U$ ou variação na configuração $PV$ do sistema. Cuja a soma define a Entalpia $H$ conforme a Eq. \ref{Eq:defentalpia} que ainda pode ser expressa de forma intensiva como a entalpia específica em base molar $\bar h = H/N$, conforme a Eq. \ref{Eq:defentalpiaespecifica}, onde $\bar u$ e $\bar v$, são a energia interna e volume ($\bar v = 1/ \bar \rho$) em base molar.

\begin{equation}
\label{Eq:defentalpia}
H = U + PV
\end{equation}

\begin{equation}
\label{Eq:defentalpiaespecifica}
\frac{H}{N} = \frac{U}{N} + P\frac{V}{N} = \bar h = \bar u + P \bar v
\end{equation}

Em sistemas complexos constituídos de múltiplas espécies de partículas, como uma mistura de gases, a entalpia da mistura $H$ pode ser obtida pela Eq. 

\begin{equation}
\label{Eq:entalpia_mis}
H^{\circ} = \sum_{i=1}^{k} N_i H_i
\end{equation}

Considerando a definição de capacidade térmica à pressão constante dada pela Eq. \ref{Eq:capacidadetermica2}, $H_i$ pode ser expresso como uma função da temperatura absoluta conforme a Eq. \ref{Eq:entalpia_esp_temp}, na qual $H^{\circ}_{i}$ é a entalpia-padrão em um estado de referência, neste texto assumida como 101325 Pa e 298,15K e $C_{p,i}$ a capacidate térmica à pressão constante do componente $i$.

\begin{equation}
\label{Eq:capacidadetermica2}
C_{P,i} = \left( \frac{\partial F_{i, mol}}{\partial T} \right)_P
\end{equation}


\begin{equation}
\label{Eq:entalpia_esp_temp}
H_i = H^{\circ}_{i} + \int_{T_{ref}}^{T} C_{P,i} dT
\end{equation}

Os valores de $h_{ref}$ geralmente são obtidos diretamente de dados e espectroscopia molecular por meio de técnicas de mecânica estatística que fogem ao escopo desta revisão.

Historicamente, tabelas termodinâmicas têm sido as fontes mais tradicionais para os valores de $H$ e $C_{P}$, no entanto, nas últimas décadas, principalmente devido à popularização de métodos numéricos aplicados à Termodinâmica, as equações polinomiais da NASA para $C_p$, $H$ e $S$ (entropia) vêm sendo empregadas, estas equações são mostradas respectivamente nas Eqs. \ref{Eq:polnasaCp, Eq:polnasaH, Eq:polnasaH}. 

\begin{equation}
\label{Eq:polnasaCp}
\frac{c_P}{R} = a_1 + a_2 T + a_3 T^2 + a_4 T^3 + a_5 T^4
\end{equation}

\begin{equation}
\label{Eq:polnasaH}
\frac{h}{R T} = a_1 + a_2 \frac{T}{2} + a_3 \frac{T^2}{3} + a_4 \frac{T^3}{4} + a_5 \frac{T^4}{5} + \frac{a_6}{T}
\end{equation}

\begin{equation}
\label{Eq:polnasaS}
\frac{s}{R} = a_1 ln(T) + a_2 T + a_3 \frac{T^2}{2} + a_4 \frac{T^3}{3} + a_5 \frac{T^4}{4} + a_7
\end{equation}

Os coeficientes $a_i$ das equações são selecionados segundo o composto, a faixa de temperatura e a pressão. 

| $CH_4$ - 1 atm | Temperatura(K): 1000 < T < 5000  | Temperatura(K): 300 < T < 1000   |
| :---   |    :----                                   |    :----                                  |
| $a_1$  |  +0.16834788E+01                           |  +0.77874148E+00                          |
| $a_2$  |  +0.10237235E−01                           |  +0.17476683E−01                          |
| $a_3$  |  −0.38751286E−05                           |  −0.27834090E−04                          |
| $a_4$  |  +0.67855849E−09                           |  +0.30497080E−07                          |
| $a_5$  |  −0.45034231E−13                           |  −0.12239307E−10                          |
| $a_6$  |  −0.10080787E+05                           |  −0.98252285E+04                          |
| $a_7$  |  +0.96233949E+01                           |  +0.13722195E+02                          |

A tabela mostra os coeficientes polinomiais para o gás metano ($CH_4$) a uma pressão de 1 atm. Em aplicações termoquímicas estes coeficientes são listados em arquivos de texto plano denominados *thermochemical data file* (TDF) - arquivos de dados termoquímicos em tradução livre - que acompanham mecanismos de reção (MR). Estes últimos são estruturas de dados mais complexas, que referenciam os parâmetros termodinâmicos nos TDFs,  os coeficientes de transporte e descrevem as reações possíveis em termos de reagentes, produtos e parâmetros cinéticos das múltiplas reações.

O GRI-Mech 3.0 é certamente o MR mais utilizado para estudos numéricos da combustão, trata-se de um mecanismo desenvolvido e otimizado para a combustão do gás natural, nele são descritas 53 espécies químicas diferentes e 325 reações, incluindo reações complexas de formação de $NO_x$ térmico e reações de combustão consecutivas. Na Fig. {} pode-se ver uma comparação dos valores de entalpia entalpia específica e capacidade térmica específica à pressão constante obtidas por três métodos distintos; as polinomiais da NASA parametrizadas pelo GRI-Mech 3.0 as tabelas de propriedades da biblioteca CoolProp  e as equações de propriedades do metano da suíte Cantera. 

```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
#plt.style.use('grayscale')
import labellines as lbl
from CoolProp.CoolProp import PropsSI
from nasa_parse import nasa_pols_ch4

#número de pontos para análise
n_points = 21

temps_kelvin = np.linspace(20,4726.85,n_points, dtype=np.float64) + 273.15

# Abordagem NASA-POLS implementado pelos autores     
vec_nasa_pol_ch4 = np.vectorize(nasa_pols_ch4) # Vetorização da função nasa_pols_ch4
entalp_nasa = vec_nasa_pol_ch4(temps_kelvin, 'H')
cp_nasa = vec_nasa_pol_ch4(temps_kelvin, "CP")

# Dados tabelados via CoolProp
entalp_coolp = PropsSI("HMOLAR", 'P', 101325, 'T', temps_kelvin, "CH4")
cp_coolp = PropsSI("CPMOLAR", 'P', 101325, 'T', temps_kelvin, "CH4")

# Abordagem equações de estado via Cantera
entalp_cantera = np.empty_like(temps_kelvin, dtype=np.float64)
cp_cantera = np.empty_like(temps_kelvin, dtype=np.float64)
gas = ct.Solution('gri30.yaml')
gas.X = "CH4:1000"
for i, t in np.ndenumerate(temps_kelvin):
    gas.TP = t, 101325
    entalp_cantera[i[0]] = gas.enthalpy_mole/1000
    cp_cantera[i[0]] = gas.cp_mole/1000

correct = entalp_coolp[0] - entalp_nasa[0]
fig, axs = plt.subplots(2,1, figsize=(6,12), facecolor='white')

axs[0].grid(which='both', alpha=.2)
axs[0].plot(temps_kelvin, (entalp_nasa + correct), 'o-.', markeredgecolor = 'k', label="$CH_4$ - NASA-pols (GRI-Mech 3.0)")
axs[0].plot(temps_kelvin, entalp_coolp, 's-.', markeredgecolor = 'k', label="$CH_4$ - CoolProp-tabs")
axs[0].plot(temps_kelvin,(entalp_cantera + correct),  'h-.', markeredgecolor = 'k', label="$CH_4$ - Cantera-eqs")
axs[0].set_ylabel("Entalpia Específica, $h$ \n ($kJ.kmol^{-1}$)")
axs[0].set_xlabel("Temperatura Absoluta \n (K)")
axs[0].legend()

axs[1].grid(which='both', alpha=.2)
axs[1].plot(temps_kelvin, cp_nasa, 'o-.', markeredgecolor = 'k',  label="$CH_4$ - NASA-pols (GRI-Mech 3.0)")
axs[1].plot(temps_kelvin, cp_coolp, 's-.', markeredgecolor = 'k',  label="$CH_4$ - CoolProp-tabs")
axs[1].plot(temps_kelvin, cp_cantera, 'h-.', markeredgecolor = 'k',  label="$CH_4$ - Cantera-eqs")
axs[1].set_xlabel("Temperatura Absoluta \n (K)")
axs[1].set_ylabel("Capacidade Térmica Específica à Pressão Constante, $c_p$ \n ($kJ.K^{-1}.kmol^{-1}$)")
axs[1].legend()
plt.show()
```

<h3>
    Abordagem Adiabática
</h3>
Como reação de natureza exotérmica, a combustão produz um montante de energia líquido proveniente dos rearranjos atômicos e moleculares próprios da oxidação do combustível. Os efeitos físicos desse exporte de energia nas proximidades da reação são inúmeros, no entanto, os mais notáveis são a tranferência de calor para a vizinhança da chama e o carreamento energético próprio da difusão do produtos de combustão, principalmente em fase gasosa.


Em uma situação ideal na qual a energia despendida pela reação não produz trabalho, não provoca variações de energia potencial e cinética, tampouco é transferida para a vizinha, o único efeito apreciável é o aquecimento do sistema, situação na qual a chama possuirá temperatura máxima para um regime permanente de combustão. 

Como durante o processo os reagentes são consumidos para a formação dos produtos, o princípio de conservação da energia aplicado da primeira lei da termodinâmica, leva a conclusão de que a entalpia dos reagentes - antes da combustão - deve ser exatemente igual à entalpia dos produtos. Conforme a Eq. \ref{Eq:conserventalpia}.

\begin{equation}
\label{Eq:conseventalpia}
\left( \sum_{i=1}^{k} N_i H_i \right)_{reagentes} = \left( \sum_{i=1}^{k} N_i H_i \right)_{produtos}
\end{equation}

Aplicando a Eq. \ref{Eq:entalpia_esp_temp} para um sistema multicomponente, onde $H_{r}$ e $H_{p}$ são respectivamente a entalpia dos reagentes  e a entalpia dos produtos em uma temperatura $T$ corresponde àquela atingida pelo aquecimento dos produtos devido ao calor de combustão, pode-se escrever a igualdade na forma dada pela Eq. \ref{Eq:entalpsigual}, de onde vem a solução analítica quando os valores $C_{P, i}(T)$ são conhecidos para os produtos.  

\begin{equation}
\label{Eq:entalpsigual}
H_{r} = H_{p} =  \sum_{i=i}^k \left( H^{\circ}_{i} \right)_{reagentes} = \sum_{i=i}^k \left( H^{\circ}_{i} + \int_{T_{ref}}^{T} C_{P,i} dT \right)_{produtos}
\end{equation}


A solução numérica no entanto é mais generalista já que $C_P(T)$ pode, dentre outras formas, ser uma polinomial da Nasa, equações de estado para os componentes de interesse ou mesmo uma função ajustada a dados tabelados.

No gráfico nota-se o ponto de igualdade entre a entalpia dos reagentes e produtos para a combustão completa do metano com oxigênio molecular como agente oxidante em um intervalo razoável de temperaturas, indicando a existência de solução para a igualdade. 

```python
from scipy.optimize import root_scalar, fsolve
from nasa_parse import nasa_pols   
def entalpia_t(temperaturas, produtos_dict, h0 = 0.):    
    # vetoriza a função nasa_pols com NumPy para ganho de desempenho
    # computacional
    vec_nasa_pol = np.vectorize(nasa_pols)
    
    h_t = np.zeros_like(temperaturas)
    for c in produtos_dict["comps"]:
        h_t += vec_nasa_pol(c[0], temperaturas, 'h') * c[1]
    return h_t - h0
 
reag_data = {
    "comps":[
        ("CH4",1), 
        ("O2",2)
    ],
}
reag_entalp = 0
for c in reag_data["comps"]:
    h_t = nasa_pols(c[0], 298.15, 'h')
    reag_entalp += c[1] * h_t

prod_data = {
    "comps":[
        ("CO2",1), 
        ("H2O",2)
    ],
}
temps_abs = np.linspace(25,3400,n_points, dtype=np.float64) + 273.15
xlim = [temps_abs[0]*.9, temps_abs[-1]*1.1]


h_t = entalpia_t(temps_abs, prod_data)

from matplotlib.offsetbox import (
    TextArea, 
    DrawingArea, 
    OffsetImage,
    AnnotationBbox
)

fig, axs = plt.subplots(figsize = (6,5), dpi=100, facecolor='white')
axs.grid(which='both', alpha=.2)

axs.plot(
    xlim, 
   [reag_entalp, reag_entalp],
    '-.', 
    markeredgecolor = 'k', 
    label = "H° - $(CH_4 + 2O_2)$"
)
axs.plot(
    temps_abs, 
    h_t, 
    'o-.', 
    markeredgecolor = 'k',  
    label="H - $(CO_2 + 2H_2 O)$"
)
axs.set_ylabel("Entalpia ($kJ$)")
axs.set_xlabel("Temperatura Absoluta (K)")
axs.set_xlim(xlim[0], xlim[1])
axs.legend()

xy = (2950, -0.05e6)
temp_adiab = fsolve(
    entalpia_t,
    args = (prod_data, reag_entalp), 
    x0 = 1000.
)
offsetbox = TextArea(f"Temperatura \nadiabática: {temp_adiab[0]:.1f}K")

ab = AnnotationBbox(offsetbox, xy,
                    xybox=(-20, 40),
                    xycoords='data',
                    boxcoords="offset points",
                    arrowprops=dict(arrowstyle="->"))
axs.add_artist(ab)


plt.show()
```

<h2>
    Equilíbrio Químico da Combustão
</h2>

Ao contrário do que geralmente se pensa, a transição entre reagentes e produtos durante uma reação química não se dá de forma direta e contínua. A formação dos produtos tem origem, na verdade, em sub-espécies mais simples provenientes da dissociação dos reagentes, como $H_2$, $OH$, $CO$, $H$, $O$, $N$, $NO$ e outros mais incomuns. Quando apreciadas sob esta óptica, a análise da temperatura adiabática de uma chama já não é mais governada apenas pelas leis da conservação de massa e energia por meio da Primeira Lei da Termodiâmica, mas inclusive pela Segunda Lei da Termodinâmica.

Suponha a combustão do metano com oxigênio como oxidante, de forma geral a reação pode ser escrita conforme a Equação . 
$$
\alpha CH_4 + \beta O_2 \rightarrow \gamma CO_2 + \mu H_2O  
$$

A entalpia de combustão máxima ocorre quando $\alpha = 0$ e $\gamma = 1$, valores intermediários, no entanto, podem ser mais provávieis, segundo fatores intínsecos e extrinsecos deslocam o equilíbrio da reação para os reagentes ou produtos. De forma geral, o balanço de massa do Carbono pode ser equacionado conforme a Equação, onde $k$ é uma constante. 

$$
\alpha + \gamma = k_C
$$
$$
2 \alpha + \mu = k_H
$$
$$
2 \beta + 2 \gamma + \mu = k_O
$$

Em um sistema ideal fechado, no qual se observe a reação de forma adiabática com volume e massa fixos, a aplicação da Segunda Lei da Termodiâmica exige que a variação da entrapia ($S$) do sistema seja restrita $dS \geq 0$. 
Ainda que a entropia de compostos isolados seja função exclusivamente da temperatura, em misturas gasosas, como é caso, a pressão parcial do sistema importa, a correção aplicada geralmente pode ser expressa na forma da Equação para gases ideais, onde $\bar s_i$ é a entropia molar. 

$$
\bar s_i = \bar s_i^\circ + \left( \Delta \bar s_i \right)_{T_{ref}}^{T} - R . ln \left( \frac{P_i}{P^\circ} \right)
$$

A Entropia da mistura pode ser obtida pelo somatório expresso na Equação. 

$$
S(T, \gamma, P_i) = \sum N_i . \bar s_i
$$

```python
from nasa_parse import nasa_pols   
    
# vetoriza a função nasa_pols co'm NumPy para ganho de desempenho
# computacional
vec_nasa_pol = np.vectorize(nasa_pols)
def entalpia_t(temperaturas, comps_dict):  
    h_t = np.zeros_like(temperaturas)
    h_0 = 0
    for c in comps_dict["comps"]:
        if c[1] == 'r':
            h_0 += vec_nasa_pol(c[0], 298.15, 'h') * c[2]
        else:
            h_t += vec_nasa_pol(c[0], temperaturas, 'h') * c[2]
    return h_t - h_0

concent = np.linspace(0,1,n_points)
comps_data = {
    "comps":[
        ("CH4", 'r',1-concent), 
        ("O2",'r', 2 * (1-concent)),
        ("CO2",'p', concent),
        ("H2O",'p',2*concent)
    ],
}

# Determina a entalpia inicial do sistema 
h_0 = 0 
for c in comps_data["comps"]:
    if c[1] == 'r':
        h_0 += vec_nasa_pol(c[0], 298.15, 'h')    

temps_abs = fsolve(
    entalpia_t,
    args = comps_data, 
    x0 = np.array([1000. for i in range(n_points)])
)

# Filtra as soluções no intervalo de temperaturas razoável
correct_vals = np.argwhere((temps_abs > 298.) & (temps_abs<5000.))
temps_abs = temps_abs[correct_vals]
concent = concent[correct_vals]
```

```python
R_NIST = 8.31446261 # J mol-1 K-1

# Refaz a tabela de componentes para ajustar
# o número de elementos ao número de temperaturas
comps_data = {
    "comps":[
        ("CH4", 'r',1-concent), 
        ("O2",'r',(4-3*concent)/2),
        ("CO2",'p', concent),
        ("H2O",'p',2*concent)
    ],
}
s_t = np.zeros_like(concent)
for idx, tt in enumerate(temps_abs):
    for comp in comps_data["comps"]:
        #if comp[1] == 'r':
        #    s_t[idx] +=  vec_nasa_pol(comp[0], 298.18, 's')*comp[2][idx]
        #else:
            s_t[idx] += vec_nasa_pol(comp[0], tt, 's')* comp[2][idx]
            s_t[idx] += vec_nasa_pol(comp[0], 298.18, 's')* comp[2][idx]
            if comp[2][idx]/3 == 0 or comp[2][idx]/3 == float('inf'):
                log_v = -1
            else:
                log_v = np.log(comp[2][idx]/3)
            
            s_t[idx] += - R_NIST * log_v * comp[2][idx]

cores = list(mpl.rcParams["axes.prop_cycle"])
mpl.rcParams["legend.framealpha"] = 1
fig, ax = plt.subplots(
    figsize = (6,5), 
    dpi=100, 
    facecolor='white'
)
ax.plot(
    concent, 
    s_t, 
    '-o', 
    color = cores[0]['color'],  
    markeredgecolor = 'k', 
    label = "Entropia"
)
ax2 = ax.twinx()
ax2.plot(
    concent, 
    temps_abs, 
    '-s', 
    color = cores[1]['color'], 
    markeredgecolor = 'k', 
    label = "Temperatura"
)

ax.set_xlabel("$CH_4$ (N° mols)")
ax.set_ylabel("Entropia (kJ/K)")
ax2.set_ylabel("Temperatura (K)")
ax.set_yscale('log')
#ax.grid(which='both', alpha=.3)
ax2.grid(which='both', alpha=.3)

#ax.plot([0.44, 0.44], [0,2.195e3], 'k-', alpha=.2)
#ax.plot([0.49, 0.49], [0,2.195e3], 'k-', alpha=.2)
#ax.fill_between([0.44, 0.49], 2.195e3,  alpha=0.2) 


ax3 = ax.twinx()
ax3.spines.right.set_position(("axes", 1.2))
ax3.plot(
    concent, 
    np.gradient(s_t.T[0], 1), 
    'h', 
    color = cores[2]['color'], 
    markeredgecolor = 'k', 
    label = "$dS/dN_{CH_4}$"
)

tkw = dict(size=4, width=1.1)
ax3.set_ylabel("$dS/dN_{CH_4}$")
ax3.tick_params(axis='y', **tkw)
ax3.axhline(y=0, color='k', linestyle='-', alpha=.25)
ax3.axvline(x=.472, color='k', linestyle='-', alpha=.25)
ax3.plot(0.47,0, 'or', markeredgecolor = 'white')
xy = [0.47,0]
offsetbox = TextArea(
    "Equilíbrio Químico\n" +\
    "\t$(T_{eq}, S_{max})$\n" +\
    "\t ≈$.472 CH_4$"
)

ab = AnnotationBbox(offsetbox, xy,
                    xybox=(60, 60),
                    xycoords='data',
                    boxcoords="offset points",
                    arrowprops=dict(arrowstyle="->"))
ax3.add_artist(ab)


fig.legend(loc='lower right')
plt.show()
```

O ponto de máxima entropia no gráfico, marca a temperatura de equilíbrio químico, consistente, como se vê, com o valor mostrado no gráfico. Neste caso, a aparente redução da entropia demostrada no gráfico, não consiste em uma violação da segunda lei, mas sim uma dedução na concentração de $CO_2$ provocada pelo equilíbrio na reação de dissociação deste composto, ou ainda por outras reações concorrentes envolvendo água e oxigênio molecular. 

$$
2CO + O_2 \Longleftrightarrow 2 CO_2
$$

A consequência imediata é uma variação na composição do sistema tendendo ao equilíbrio químico pela esquerda e direita de uma concentração de $CO_2$ dada por $(0,4 < c < 0,6) \frac{mols_{CO_2}}{mols_{CH4}}$, onde o ponto exato de equilíbrio é definido como $dS = 0$.

Para uma análise mais pormenorizada do fenômeno, geralmente o equilíbrio da reação é estudado por meio da Energia Livre de Gibbs $G$, definida pela Equação.

$$
G \equiv H - TS
$$

donde vem uma segunda expressão para a segunda lei dada por $dG \leq 0$, com $dG = 0$ no ponto de equilíbrio químico, na qual os valores $\bar g_{i,T}$ para cada espécie podem ser escritos na forma da Equação, tomando o modelo de gás ideal.

$$
\bar g_{i,T} = \bar g_{i,T}^{o} + R_u T ln \left( \frac{P_i}{P^{o}} \right)
$$

Com o a energia livre de Gibbs da mistura dada pela Equação

$$
G_{mix} = \sum N_i \bar g_{i,T}
$$

Assim é possível inferir que no equilíbrio químico de uma reação formada por $i$ compostos, dentre reagentes e produtos, e considerando a segunda lei da termodinâmica, a variação de $G_{mix}$ deve ser nula

$$
dG_{mix} = \sum dN_i \left[ \bar g_{i,T}^{o} + R_u T ln \left( \frac{P_i}{P^{o}} \right) \right] = 0
$$

Se um sistema genérico for considerado, por exemplo $aA + bB + cC \leftrightarrow dD + eE + fF$, então a variação de $dN$ de cada componente é diretamente proporcional ao coeficiente estequimétrico da reação por meio de uma razão $\kappa < 0$ para os reagentes, já que são consumidos na reação  $\kappa > 0$ para os produtos.

$$
\begin{matrix} 
dN_A = - \kappa a \\ 
dN_B = - \kappa b \\
\vdots \qquad \vdots \\
dN_D = - \kappa d \\ 
dN_E = - \kappa e \\
\vdots \qquad \vdots
\end{matrix}$$

Neste caso, a variação da energia livre de Gibbs para uma reação pode ser esquacionada conforme a Equação, onde cada parcela tem a forma $\pm i [\bar g^{o}_{A, T} + RT ln(P_I / P^o)]$, com $\kappa$ definindo conforme a Equação. 

$$
\Delta G_T^o  = \left( \sum p . \bar g^{o}_{P,T} \right)_{produtos} - \left( \sum r . \bar g^{o}_{R,T} \right)_{reagentes}
$$

$$
\kappa = \bar g^{o}_{i, T} + RT ln\left( \frac{P_i}{P^o} \right)
$$

Para o caso do Equilíbrio químico é possível demonstrar que a constante de equilíbrio para a reação $\kappa_P$ assume a forma da Equação

$$
\kappa_P = exp \left( \frac{\Delta G_T^o}{RT} \right)
$$




Conhecidos os valores de $\kappa_P$ e $\Delta G_{T}^{o}$, o comportamento da reação pode ser apreciado qualitativamente, pelos seguintes critérios; para $\Delta G_{T}^{o} > 0 $ e $\kappa_P < 1$ a reação tende à reconstituição dos compostos reagentes por meio de reações mais simples, se $\Delta G_{T}^{o} < 0$, então tende aos produtos. Aleḿ disso, com $\kappa_P > 1$, e a reação tendendo aos produtos, a reação será também exotérmica, já que a definição de $G$, exige que a variação da entalpia seja negativa. 

Deste modo, a partir da definição de temperatura adiabática, quanto da definição de equilíbrio químico é possível prever simultaneamente, tanto a temperatura da chama ao longo da reação, quanto a constituição do sistema em termos da fração residual de reagentes e constituição dos produtos esperados.
