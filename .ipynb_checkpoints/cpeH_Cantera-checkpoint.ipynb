Apesar da mutabilidade das sintaxes de linguagens de programação e principalmente dos métodos de uso de uma determinada biblioteca devido ao processo de evolução e obsolecência das features, os autores, acreditam que a exposição da listagem abaixo pode ser extensivamente instrutiva para um grande número de leitores por um intervalo de tempo considerável. 

Nela se vê o processo de construção de gráficos da capacidade térmica à pressão constate e entalpia em função da temperatura.

Além de Cantera, foram ainda utilizadas a biblioteca NumPy, para a manipulação de vetores com alto desempenho computacional e o módulo PyPlot da biblioteca Matplotlib para a construção dos gráficos. 

```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('grayscale')
```

A modelagem de um gás através da suite Cantera se dá através da criação de um objeto do tipo \texttt{Solution()}, neste caso, foi selecionado o MR GRI-Mech 3.0 e ao objeto atributos de temperatura como 298,15 K, pressão como 101325 Pa e constituição molar de 1 mol de metano. 

```python
gas = ct.Methane()
gas.TP = 298.15, 101325
gas()
```

Apesar de ser método de desempenho computacional-numérico inferior, a forma mais intuitiva de obter os ditos diagramas é iterando em laços aninhados sobre vetores previamente inicializados.

```python
temps_kelvin = np.linspace(0,700,100, dtype=np.float64) + 273.15
press_pascal = np.logspace(0.01,5,40, dtype=np.float64) * 101325
entalpia = np.empty_like(temps_kelvin, dtype=np.float64)
cap_c_press = np.empty_like(temps_kelvin, dtype=np.float64)

fig, axs = plt.subplots(1,2, figsize=(16,5), facecolor='white')
for p in press_pascal:
    for i, t in np.ndenumerate(temps_kelvin):
        gas.TP = t, p
        entalpia[i[0]] = gas.enthalpy_mole
        cap_c_press[i[0]] = gas.cp_mole
    axs[0].plot(entalpia, temps_kelvin)
    axs[1].plot(temps_kelvin, cap_c_press)
axs[0].set_xlabel("Entalpia \n (kJ/kmol)")
axs[0].set_ylabel("Temperatura Absoluta \n (K)")
axs[0].set_xscale("log")
axs[1].set_xlabel("Temperatura Absoluta \n (K)")
axs[1].set_ylabel("Capacidade Térmica à Pressão Constante \n ($J.K^{-1}.kmol^{-1}$)")
plt.show()
```
